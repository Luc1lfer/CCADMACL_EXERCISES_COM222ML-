# Exercise 4

## Import the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import normalize
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
from sklearn.neighbors import NearestNeighbors
```

## Load the data

```python
X = pd.read_csv('creditcard.csv')

# Dropping the CUST_ID column from the data
X = X.drop('CUST_ID', axis = 1)

# Handling the missing values
X.fillna(method ='ffill', inplace = True)

print(X.head())
```

## Pre-process the data

```python
# Scaling the data to bring all the attributes to a comparable level
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Normalizing the data so that
# the data approximately follows a Gaussian distribution
X_normalized = normalize(X_scaled)

# Converting the numpy array into a pandas DataFrame
X_normalized = pd.DataFrame(X_normalized)
```

## Reduce the dimensionality of the data using PCA

```python
pca = PCA(n_components = 2)
X_principal = pca.fit_transform(X_normalized)
X_principal = pd.DataFrame(X_principal)
X_principal.columns = ['P1', 'P2']
print(X_principal.head())
```

```python
nbrs = NearestNeighbors(n_neighbors=5).fit(X_principal)
distances, indices = nbrs.kneighbors(X_principal)
distances = np.sort(distances[:, 4])
plt.figure(figsize=(8, 5))
plt.plot(distances)
plt.title('k-distance Graph')
plt.xlabel('Data Points sorted by distance')
plt.ylabel('Epsilon (eps)')
plt.show()
```

## Train a DBSCAN model using default values

```python
# Numpy array of all the cluster labels assigned to each data point
db_default = DBSCAN().fit(X_principal)
labels = db_default.labels_
```

## Plot the clustered data

```python
# Building the label to colour mapping
colours = {}
colours[0] = 'r'
colours[1] = 'g'
colours[2] = 'b'
colours[-1] = 'k'

# Building the colour vector for each data point
cvec = [colours[label] for label in labels]

# For the construction of the legend of the plot
r = plt.scatter(X_principal['P1'], X_principal['P2'], color ='r');
g = plt.scatter(X_principal['P1'], X_principal['P2'], color ='g');
b = plt.scatter(X_principal['P1'], X_principal['P2'], color ='b');
k = plt.scatter(X_principal['P1'], X_principal['P2'], color ='k');

# Plotting P1 on the X-Axis and P2 on the Y-Axis
# according to the colour vector defined
plt.figure(figsize =(9, 9))
plt.scatter(X_principal['P1'], X_principal['P2'], c = cvec)

# Building the legend
plt.legend((r, g, b, k), ('Label 0', 'Label 1', 'Label 2', 'Label -1'))

plt.show()
```

1. Train a DBSCAN model

```python
# put your answer here
db = DBSCAN(eps=0.41, min_samples=2000).fit(X_principal)
labels = db.labels_
```

2. Plot the clustered data

```python
#put your answer here
colours = {0: 'r', 1: 'g', 2: 'b', -1: 'k'}
cvec = [colours[label] if label in colours else 'y' for label in labels]

plt.figure(figsize=(9, 9))
plt.scatter(X_principal['P1'], X_principal['P2'], c=cvec)
plt.title('DBSCAN Clustering')
plt.xlabel('P1')
plt.ylabel('P2')
plt.show()
```

3. Print the number of clusters

```python
#put your answer here
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
print(f'Number of clusters: {n_clusters}')
```

4. Print the number of noisy data

```python
#put your answer here
n_noise = list(labels).count(-1)
print(f'Number of noisy data points: {n_noise}')
```

5. Print the silhouette score

```python
#put your answer here
if n_clusters > 1:  # Silhouette score requires at least 2 clusters
    score = silhouette_score(X_principal, labels)
    print(f'Silhouette Score: {score}')
else:
    print('Silhouette Score: Cannot be calculated with less than 2 clusters')
```
