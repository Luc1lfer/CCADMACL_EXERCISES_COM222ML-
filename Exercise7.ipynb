```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.preprocessing import TransactionEncoder
```

```python

#reading data
url = "https://github.com/robitussin/CCADMACL_EXERCISES/blob/ff41fec98bd270f4bbeedb9b8849ae21fdeadfee/Groceries_dataset.csv?raw=true"
dataset = pd.read_csv(url)

dataset.head()
```

```python
client_basket = dataset.groupby(['Member_number', 'Date'])['itemDescription'].apply(lambda x: ', '.join(x))
client_basket
```

### 1. Apply one-hot encoding

```python
## put your answer here
one_hot_transformer = TransactionEncoder()
df_transform = one_hot_transformer.fit_transform(client_basket.str.split(', '))
df_transform
```

```python
df_encoded = pd.DataFrame(df_transform,columns=one_hot_transformer.columns_)
df_encoded
```

2. Use the Apriori Algorithm to find the most frequent sets

```python
## put your answer here
df_ap = apriori(df_encoded, min_support = 0.001, use_colnames = True) # Lowering the min_support
df_ap.sort_values(['support'],ascending=False, inplace = True)
```

3. Display the association rules using the lift metric

```python
## put your answer here
df_ar = association_rules(df_ap, metric="lift", min_threshold=1)
df_ar
```
